```python
import google.generativeai as genai
from google.colab import userdata
genai.configure(api_key=userdata.get('GEMINI_API_KEY'))
```

# 1. Blog generator

```python
def generate_blog(topic, tone, word_limit=60):
    system_prompt = (
        "You are a professional blog writer. "
        "Write engaging, informative, and easy-to-read blogs."
    )
    user_prompt = f"""
    Write a blog post.
    Topic: {topic}
    Tone: {tone}
    Length: approximately {word_limit} words
    The blog should have:
    - A catchy title
    - Introduction
    - Main content
    - Conclusion
    """

    model = genai.GenerativeModel(
        model_name='gemini-2.5-flash-lite',
        system_instruction=system_prompt
    )
    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.8,
            "max_output_tokens": 300
        }
    )
    return response.text

blog_topic = input("Enter Blog Topic: ")
blog_tone = input("Enter Blog Tone (Professional/Friendly/Casual): ")
blog_length = input("Enter Blog Length in words: ")
print("\nGenerating Blog...\n")
print(
    generate_blog(
        blog_topic,
        blog_tone,
        blog_length
    )
)
```

# 2. Email generator

```python
def generate_email(purpose, tone, recipient, sender):
    system_prompt = (
        "You are an expert email writer. "
        "Write clear, professional, and polite emails."
    )
    user_prompt = f"""
    Write an email.
    Purpose: {purpose}
    Tone: {tone}
    Recipient Name: {recipient}
    Senders Name: {sender}
    The email should include:
    - Subject line
    - Greeting
    - Body
    - Closing
    """

    model = genai.GenerativeModel(
        model_name='gemini-2.5-flash-lite',
        system_instruction=system_prompt
    )
    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.7,
            "max_output_tokens": 400
        }
    )
    return response.text

email_purpose = input("Enter Email Purpose: ")
email_tone = input("Enter Email Tone: ")
sender = input("Enter Senders Name: ")
recipient = input("Enter Recipient Name: ")
print("\nGenerating Email...\n")
print(
    generate_email(
        email_purpose,
        email_tone,
        recipient,
        sender
    )
)
```

```python
import google.generativeai as genai
from google.colab import userdata
API_KEY = userdata.get("GEMINI_API_KEY")
genai.configure(api_key=API_KEY)
```

```python
model = genai.GenerativeModel(
    model_name="gemini-2.5-flash-lite",
    generation_config={
        "temperature": 0.5,
        "max_output_tokens": 500
    }
)
system_prompt = """
You are an AI-powered Travel and Tourism FAQ Assistant.
Your job is to help users with:
- tourist destinations
- travel planning
- hotel suggestions
- food recommendations
- travel budgets
- itineraries
- transportation
- safety tips
- weather guidance

Rules:
1. Be friendly and professional.
2. Keep answers clear and concise.
3. Suggest practical travel advice.
4. If you do not know something, say politely that you are unsure.
5. Use emojis occasionally to make responses engaging.
"""

faq_data = {
    "best places in kerala":
    """
    🌴 Top tourist places in Kerala:
    1. Munnar
    2. Alleppey
    3. Wayanad
    4. Varkala
    5. Thekkady
    """,
    "best time to visit goa":
    """
    ☀️ The best time to visit Goa is from November to February.
    The weather is pleasant and ideal for beaches and nightlife.
    """,
    "budget trip to manali":
    """
    💰 A budget Manali trip can cost around ₹8,000–₹15,000
    including travel, stay, and food for 3 days.
    """,
    "famous food in kerala":
    """
    🍲 Famous Kerala foods:
    - Appam and Stew
    - Kerala Sadya
    - Puttu and Kadala Curry
    - Malabar Biryani
    - Karimeen Pollichathu
    """,
    "places to visit in goa":
    """
    🏖️ Popular places in Goa:
    - Baga Beach
    - Calangute Beach
    - Fort Aguada
    - Dudhsagar Falls
    - Anjuna Beach
    """
}

chat = model.start_chat(history=[])
def tourism_chatbot(user_input):
    query = user_input.lower()
    for faq in faq_data:
        if faq in query:
            return faq_data[faq]

    full_prompt = f"""
    {system_prompt}
    User Question:
    {user_input}
    """
    response = chat.send_message(full_prompt)
    return response.text
```

```python
print("🌍 Welcome to AI Travel & Tourism FAQ Chatbot!")
print("Type 'exit' to stop the chatbot.\n")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("\nBot: Thank you for using the Travel Chatbot ✈️")
        break

    answer = tourism_chatbot(user_input)
    print("\nBot:", answer)
    print()
```
